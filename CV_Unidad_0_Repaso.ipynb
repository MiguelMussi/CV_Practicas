## Representación de imágenes con Numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.zeros((100, 200, 3), dtype=np.uint8)
image
```

```python
image[:50, :100] = [255, 0, 0]   # Rojo
image
```

```python
image[:50, 100:] = [0, 255, 0]   # Verde
image
```

```python
image[50:, :100] = [0, 0, 255]   # Azul
image
```

```python
image[50:, 100:] = [100, 100, 100]   # Blanco
image
```

```python
plt.figure(figsize=(4,4))
plt.imshow(image)
plt.title("Imagen Original")

plt.xticks(np.arange(0, image.shape[1]+1, 25))
plt.yticks(np.arange(0, image.shape[0]+1, 20))

plt.show()
```

```python
print(f"Shape de la imagen: {image.shape}")
```

```python
R, G, B = image[:,:,0], image[:,:,1], image[:,:,2]
```

```python
fig, axs = plt.subplots(1, 3, figsize=(10, 4))

axs[0].imshow(R, cmap='Reds')
axs[0].set_title('Canal Rojo')


axs[1].imshow(G, cmap='Greens')
axs[1].set_title('Canal Verde')

axs[2].imshow(B, cmap='Blues')
axs[2].set_title('Canal Azul')

for ax in axs:
  ax.set_xticks(np.arange(0, image.shape[1]+1, 50))
  ax.set_yticks(np.arange(0, image.shape[0]+1, 25))

plt.show()
```

## Imagen desde URL

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import requests
from io import BytesIO
```

```python
# Función auxiliar para leer una imagen desde una URL con OpenCV
def read_image_from_url(url):
    response = requests.get(url)
    image = np.array(bytearray(response.content), dtype=np.uint8)
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)  # Convertir bytes a imagen
    return image
```

```python
# URL de la imagen que deseas descargar
image_url = "https://raw.githubusercontent.com/jpmanson/tuia-unr/main/images/big_buck_bunny.jpg"

# Leer la imagen
image2 = read_image_from_url(image_url)
```

```python
image2
```

```python
# OpenCV maneja imágenes en formato BGR, convertir a RGB
image_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
```

```python
# Mostramos la información de la imagen
print(f"Shape de la imagen: {image_rgb.shape}.")
print(f"Columnas (Ancho): {image_rgb.shape[1]}")
print(f"Filas (Altura): {image_rgb.shape[0]}")
print(f"Canales: {image_rgb.shape[2]}")
print(f"Tipo de dato del array: {image_rgb.dtype}")
```

```python
plt.figure(figsize=(15,5))
plt.imshow(image_rgb)
plt.title("Imagen Original")

plt.xticks(np.arange(0, image_rgb.shape[1], 200))
plt.yticks(np.arange(0, image_rgb.shape[0], 100))

plt.show()
```

```python
# Descomponer la imagen en los canales R, G, B
R, G, B = image_rgb[:,:,0], image_rgb[:,:,1], image_rgb[:,:,2]
```

```python
# Mostrar cada canal de color
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(R, cmap='Reds')
axs[0].set_title('Canal Rojo')
# axs[0].axis('off')

axs[1].imshow(G, cmap='Greens')
axs[1].set_title('Canal Verde')
axs[1].axis('off')

axs[2].imshow(B, cmap='Blues')
axs[2].set_title('Canal Azul')
axs[2].axis('off')

plt.show()
```

## Librería Pillow

```python
import requests
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from io import BytesIO
```

```python
# URL de la imagen que deseas descargar
image_url = "https://raw.githubusercontent.com/jpmanson/tuia-unr/main/images/big_buck_bunny.jpg"
```

```python
# Paso 2: Descargar la imagen
response = requests.get(image_url)
image3 = Image.open(BytesIO(response.content))
```

```python
# Paso 3: Convertir la imagen en un array de Numpy
image_np = np.array(image3)
print(f"Shape de la imagen: {image_np.shape}.")
print(f"Columnas (Ancho): {image_np.shape[1]}")
print(f"Filas (Altura): {image_np.shape[0]}")
print(f"Canales: {image_np.shape[2]}")
print(f"Tipo de dato del array: {image_np.dtype}")
```

```python
# Paso 4: Descomponer la imagen en los canales R, G, B
R, G, B = image_np[:,:,0], image_np[:,:,1], image_np[:,:,2]
```

```python
# Paso 5: Mostrar cada canal de color
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(R, cmap='Reds')
axs[0].set_title('Canal Rojo')
axs[0].axis('off')

axs[1].imshow(G, cmap='Greens')
axs[1].set_title('Canal Verde')
# axs[1].axis('off')

axs[2].imshow(B, cmap='Blues')
axs[2].set_title('Canal Azul')
axs[2].axis('off')

plt.show()
```

## Escala de grises

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import requests
from io import BytesIO
```

```python
# Función para leer una imagen desde una URL
def read_image_from_url(url):
    response = requests.get(url)
    image = np.array(bytearray(response.content), dtype=np.uint8)
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)  # Convertir bytes a imagen
    return image
```

```python
# URL de la imagen que vamos a convertir a escala de grises
image_url = "https://raw.githubusercontent.com/jpmanson/tuia-unr/main/images/big_buck_bunny.jpg"

# Leer la imagen
image_color = read_image_from_url(image_url)
```

```python
# Convertir de BGR a RGB para mostrar correctamente con Matplotlib
image_rgb = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)
```

```python
# Convertir la imagen a escala de grises
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
```

```python
# Imprimir el shape y el tipo de dato del array de la imagen original y en escala de grises
print(f"Shape de la imagen original (RGB): {image_rgb.shape}")
print(f"Tipo de dato del array de la imagen original: {image_rgb.dtype}")
print(f"Shape de la imagen en escala de grises: {image_gray.shape}")
print(f"Tipo de dato del array de la imagen en escala de grises: {image_gray.dtype}")
```

```python
# Mostrar ambas imágenes, original y en escala de grises
plt.figure(figsize=(12, 6))

# Imagen original
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Imagen Original")
plt.axis('off')

# Imagen en escala de grises
plt.subplot(1, 2, 2)
plt.imshow(image_gray, cmap='gray')
plt.title("Imagen en Escala de Grises")
plt.axis('off')

plt.show()
```
