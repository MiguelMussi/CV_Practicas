# 6) Localización vs Clasificación

La localización de objetos y la clasificación de imágenes son dos tareas fundamentales en el campo de la visión por computadora, pero tienen objetivos y enfoques diferentes:

1. **Localización de Objetos**:
   - La localización de objetos implica detectar la presencia de objetos específicos en una imagen y también identificar su ubicación precisa dentro de la imagen.
   - El objetivo principal es encontrar las coordenadas del cuadro delimitador (bounding box) que rodea a cada objeto de interés en la imagen.
   - Generalmente, se utilizan algoritmos de detección de objetos que pueden identificar múltiples objetos en una imagen y proporcionar las coordenadas de los cuadros delimitadores.
   - Ejemplo: En una imagen que contiene diferentes tipos de frutas, la localización de objetos identificaría cada fruta individualmente y proporcionaría un cuadro delimitador para cada una, indicando su ubicación dentro de la imagen.

2. **Clasificación de Imágenes**:
   - La clasificación de imágenes se centra en asignar una etiqueta o categoría a una imagen completa, sin necesidad de identificar la ubicación específica de los objetos dentro de la misma.
   - El objetivo es identificar y clasificar el contenido general de la imagen en una categoría predefinida.
   - Por lo general, se utilizan algoritmos de clasificación de imágenes que han sido entrenados con una amplia variedad de imágenes etiquetadas previamente.
   - Ejemplo: En una colección de imágenes de animales, la clasificación de imágenes identificaría si una imagen muestra un perro, un gato, un pájaro, etc., pero no proporcionaría información sobre dónde están ubicados los animales en la imagen.

Para ilustrar la diferencia entre ambas tareas, consideremos el siguiente ejemplo:

Imaginemos que tenemos una imagen que muestra un jardín con varios perros y gatos jugando.

- La **localización de objetos** identificaría y delimitaría cada perro y gato en la imagen, proporcionando cuadros delimitadores alrededor de cada uno para indicar su ubicación exacta en la imagen.

- La **clasificación de imágenes** simplemente determinaría que la imagen es de un jardín con perros y gatos, sin preocuparse por identificar la ubicación exacta de cada animal dentro de la imagen.

En este caso, la localización de objetos nos daría información detallada sobre dónde están ubicados cada perro y gato en la imagen, mientras que la clasificación de imágenes solo nos diría el contenido general de la imagen, es decir, que contiene perros y gatos en un jardín.

En resumen, mientras que la localización de objetos se enfoca en identificar y ubicar objetos específicos dentro de una imagen, la clasificación de imágenes se centra en asignar una etiqueta o categoría general a la imagen en su conjunto.

```python
import urllib.request
from PIL import Image
from io import BytesIO

# url = 'https://miro.medium.com/v2/resize:fit:640/1*Mj8WKVKf_RpiAsX3SC1ZdQ.png'
url = 'https://dev-to-uploads.s3.amazonaws.com/uploads/articles/7qgc1cav8ch27bdqxo9n.png'

# Crear una solicitud con un encabezado de agente de usuario (Por restricciones de acceso)
req = urllib.request.Request(url, headers={'User-Agent': 'Mozilla/5.0'})
response = urllib.request.urlopen(req)
img = Image.open(BytesIO(response.read()))
img
```
